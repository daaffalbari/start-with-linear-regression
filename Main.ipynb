```python
%matplotlib widget

import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

### Generate Data

```python
jumlah_data = 100

y = np.array([i*0.1+np.random.randn() for i in range(jumlah_data)])
x = np.array([i*0.1 for i in range(jumlah_data)])
```

### Visualisasi Data

```python
plt.scatter(x,y)
plt.show()
```

### Fungsi Linear

```python
def funcLinear(x, gradien):
  y = gradien*x
  return y
```

```python
x_prediction = np.array([0, 10])
m_init = 1
y_prediction = funcLinear(x_prediction, m_init)
print(x_prediction)
print(y_prediction)
```

Menghitung Prediksi

```python
m_prediction = m_init
m_list_prediction = []
y_list_prediction = []
x_list_prediction = []
learning_rate = 0.1

for i in range(1, jumlah_data):
  y_prediction = funcLinear(x[i], m_prediction)
  y_actual = y[i]
  
  error = y_actual - y_prediction
  delta_m = learning_rate*error/x[i]
  m_prediction = m_prediction + delta_m
  
  m_list_prediction.append(m_prediction)
  y_list_prediction.append(funcLinear(np.array([0, 10]), m_prediction))
  x_list_prediction.append(np.array([0, 10]))
```

Visualisasi Prediksi

```python
fig = plt.figure()
line, = plt.plot([], [], "RED")

def animate(frame_num):
  x = x_list_prediction[frame_num]
  y = y_list_prediction[frame_num]
  line.set_data(x, y)
  return line, 

anime = FuncAnimation(fig, animate, frames=100, interval=100, repeat=False)
plt.scatter(x,y)
plt.axis([0, 10, 0, 10])
plt.show()
```
